# Automação de Calculo de Faturamento
#### Essa automação baixa um dataset, calcula o faturamente e o total de produtos vendidos e manda essas informções por E-mail.

```python
!pip install pyautogui
!pip install pyperclip
!pip install pandas

import pyautogui as pgui
import pyperclip
import time
import pandas as pd
import sys
from email.message import EmailMessage 
import smtplib

pgui.PAUSE=1
pgui.hotkey('ctrl', 't') 
pyperclip.copy('') #Link for download the dataset
pgui.hotkey('ctrl', 'v')
pgui.press('Enter')
time.sleep(5)
pgui.click(x=520, y=417, clicks=1)
pgui.click(x=1652, y=192, clicks=1)
pgui.click(x=1464, y=619, clicks=1)

time.sleep(5)
try:
    file = pd.read_excel(r'') #Path of download file for open
    display(file)
except:
    print('Esse Arquivo não foi Baixado...')
    exit(0)

Lucro_total = 0
Total_Vendida = 0
for pu, qv in zip(file['Preco(por unidade)'], file['Qunt.Vendida']):
    Lucro_total = Lucro_total + (pu * qv)

Total_Vendida = file['Qunt.Vendida'].sum()


ADDRESS = '' #Your Address
PASSWORD = '' #Your Password
mail = EmailMessage()
mail['Subject'] = '#1 - Faturamento Mensal - 18/Jan/20**'
mail['From'] = ADDRESS
mail['To'] = '' #Email to
mail.set_content(f'Lucro Total do mês: R$ {Lucro_total} Reais, Quantidade de produtos que foram vendidos em todas as lojas: {Total_Vendida} Itens')

with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtpp:
    smtpp.login(ADDRESS,PASSWORD)
    smtpp.send_message(mail)
    
```
